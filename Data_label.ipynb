```python
data_label = data2_df[data2_df['type'] != 'unknown']
```

```python
print("\nDataFrame Setelah Menghapus Baris dengan 'unknown':")
print(data_label)
```

```python
def label(row):
  ph = row['ph']
  temp = row['temp']
  type = row['type']

  if type == 'ikan nila' and 6.5 <= ph <= 8.5 and 25 <= temp <= 30:
    return 'optimal'

  elif type == 'ikan lele' and 6.5 <= ph <= 8.5 and 26 <= temp <= 30:
    return 'optimal'

  elif type == 'ikan gurame' and 6.5 <= ph <= 8 and 26 <= temp <= 30:
    return 'optimal'

  else:
    return 'kurang optimal'
```

```python
data_label['label'] = data_label.apply(label, axis=1)
```
