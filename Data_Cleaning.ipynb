# 5. Menggabungkan dataset

```python
realfish_df_filtered = realfish_df[['ph', 'temperature']].rename(columns={'ph': 'pH', 'temperature': 'Temp'})
realtime_df_filtered = realtime_df[['PH', 'TEMP']].rename(columns={'PH': 'pH', 'TEMP': 'Temp'})
```

```python
data_df = pd.concat([realfish_df_filtered, realtime_df_filtered], ignore_index=True)
```

Mengubah format menjadi pembulatan

```python
data_df['Temp'] = data_df['Temp'].round(1)
```

```python
data_df['pH'] = data_df['pH'].round(1)
```

# 6. Menyimpan Dataset

```python
save_dir = '/mnt/data'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
file_path = os.path.join(save_dir, 'data.csv')
```

```python
data_df.to_csv(file_path, index=False)

print(f"Dataset gabungan telah disimpan ke {file_path}")
```
